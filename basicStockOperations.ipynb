Yahoo query is a wrapper for yahoo finance, install it using pip
https://github.com/dpguthrie/yahooquery

Getting price history for stock based on ticker
Go to Yahoo finance website, type company name and get the ticker
https://in.finance.yahoo.com/

```python
from yahooquery import Ticker
tick = Ticker('LT.NS ', asynchronous=True)
#For fetching data through multiple tickers
#tick = Ticker('LT.NS ICICIBANK.NS', asynchronous=True)
```

Getting brief summary of a stock

```python
summary = tick.summary_detail
summary
```

```python
history = tick.history()
history
```

```python
income_statement = tick.income_statement()
income_statement
```
